```python
import pandas as pd#loading the required packages
import numpy as np
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn.metrics import classification_report as rep
from sklearn.metrics import confusion_matrix as mat
from sklearn import metrics as met
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
import seaborn as sea
import pydotplus
import matplotlib
import matplotlib.pyplot as plot
from IPython.display import Image  
```

```python
mturk_sample=pd.read_csv('../data/mturk_sample.csv',index_col=0)#loading the mturk sample from the data folder
mturk_sample['pos'] = np.where(mturk_sample['class']=='pos', 1.0, 0)
#creating a pos column which has value 1 if class is equal to pos and otherwise 0
mturk_sample['neg'] = np.where(mturk_sample['class']=='neg', 1.0, 0)
#creating a neg column which has value 1 if class is equal to neg and otherwise 0
```

```python
mturk_sample[['class','annotator','pos','neg']].head()
#showing the output of the desired columns 
```

```python
mturk_sample_agg=mturk_sample.groupby(['id']).sum()
#grouping based on the id and then adding the positive and negative counts per movie thereby as a result of grouping
mturk_sample_agg['class']=np.where(mturk_sample_agg['pos']>mturk_sample_agg['neg'],'pos','neg')
mturk_sample_agg[['pos','neg','class']].head()
```

```python
xtrain=mturk_sample_agg.drop(['class','pos','neg'],axis=1)
#getting the xtrain dataframe as all the columns other than the class column and pos and neg inorder to match the test dataset
ytrain=mturk_sample_agg['class']
#obtaining ytrain as the values of the class present based on majority voting
```

```python
dtree=dt()#assigning the decision tree classifier 
dtree.fit(xtrain,ytrain)#fitting the decision tree classifier model based on the x train and y train values
```

```python
test=pd.read_csv('../data/test.csv',index_col=0)#uploading the test dataset from the data folder
xtest=test.drop(['class'],axis=1)#getting the xtest dataframe as all the columns other than the class column of the test dataset
ytest=test['class']#obtaining ytrain as the values of the class present for each id row
ypredict=dtree.predict(xtest)#predicting the values of y based on the xtest values using the model developed
```

```python
print(mat(ytest,ypredict))#printing the confusion matrix for the values of y test and y predict
print(rep(ytest,ypredict))#providing the report for the comparison between y test and y predict
print("accuracy",met.accuracy_score(ytest,ypredict)*100) #getting the accuracy
print("F1 value is ",met.f1_score(ytest,ypredict,average= 'micro'))# providing the f1 value
print("Recall value is",met.recall_score(ytest,ypredict,average='binary',pos_label="neg"))#providing the recall value for neg
print("Recall value is",met.recall_score(ytest,ypredict,average='binary',pos_label="pos"))#providing the recall value for pos
print("Predicted probabilites are",dtree.predict_proba(xtrain))#printing the predicted probabilities for x train
```
