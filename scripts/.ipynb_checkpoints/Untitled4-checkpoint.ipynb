```python
import pandas as pd#loading the required packages
import numpy as np
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn.metrics import classification_report as rep
from sklearn.metrics import confusion_matrix as mat
from sklearn import metrics as met
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
import seaborn as sea
import pydotplus
import matplotlib
import matplotlib.pyplot as plot
from IPython.display import Image  
```

```python
mturk_sample=pd.read_csv('../data/mturk_sample.csv')
mturk_sample.head()
```

```python
temp=mturk_sample[['id','annotator','class']]
temp=temp.pivot(index='id',columns='annotator',values='class')
temp.head()
```

```python
cols = list(temp.columns.values)
cols
```

```python
temp[cols] = temp[cols].replace({np.nan:0, 'pos':1,'neg':-1})
```

```python
temp.head()
```

```python
a=pd.DataFrame(temp[cols].sum())
a
```

```python
temp1=temp.groupby(['id']).sum()
```

```python
temp1.head()
```

```python
x=temp1.index
for i in x:
    temp1['clas']=0
temp1.head()
```

```python
temp1['class']=temp1.sum(axis=1)
temp1['class']=np.where(temp1['class']>=0,'pos','neg')
temp1['class'].head()
```

```python
mturk_sample_agg=mturk_sample.groupby(['id']).sum()
mturk_sample_agg['class']=temp1['class']
#grouping based on the id and then adding the positive and negative counts per movie thereby as a result of grouping
mturk_sample_agg['class'].head()
```

```python
xtrain=mturk_sample_agg.drop(['class'],axis=1)
#getting the xtrain dataframe as all the columns other than the class column inorder to match the test dataset
ytrain=mturk_sample_agg['class']
#obtaining ytrain as the values of the class present based on Dawid Skene
```

```python
dtree=dt()#assigning the decision tree classifier 
dtree.fit(xtrain,ytrain)#fitting the decision tree classifier model based on the x train and y train values
```

```python
test=pd.read_csv('../data/test.csv',index_col=0)#uploading the test dataset from the data folder
xtest=test.drop(['class'],axis=1)#getting the xtest dataframe as all the columns other than the class column of the test dataset
ytest=test['class']#obtaining ytrain as the values of the class present for each id row
ypredict=dtree.predict(xtest)#predicting the values of y based on the xtest values using the model developed
```

```python
print(mat(ytest,ypredict))#printing the confusion matrix for the values of y test and y predict
print(rep(ytest,ypredict))#providing the report for the comparison between y test and y predict
print("accuracy",met.accuracy_score(ytest,ypredict)*100) #getting the accuracy
print("F1 value is ",met.f1_score(ytest,ypredict,average= 'micro'))# providing the f1 value
print("Recall value is",met.recall_score(ytest,ypredict,average='binary',pos_label="neg"))#providing the recall value for neg
print("Recall value is",met.recall_score(ytest,ypredict,average='binary',pos_label="pos"))#providing the recall value for pos
print("Predicted probabilites are",dtree.predict_proba(xtrain))#printing the predicted probabilities for x train
```

```python
#now making these results written into train_ds.txt
con=mat(ytest,ypredict)#storing the confusion matrix values for y test and y predict
rep=rep(ytest,ypredict)# storing the report for it
acc=met.accuracy_score(ytest,ypredict)*100#storing the accuracy of the prediction
f1=met.f1_score(ytest,ypredict,average= 'micro')#storing the f1 values
posrec=met.recall_score(ytest,ypredict,average='binary',pos_label="pos")#storing the pos recall score
negvec=met.recall_score(ytest,ypredict,average='binary',pos_label="neg")#storing the neg recall score
pred=dtree.predict_proba(xtrain)#storing the prediction probability for the x train
with open('../results/train_ds.txt','w') as i: #writing a text file into the results folder as mentioned in the question
    i.write("The confusion matrix generated using dawid skene is: \n %s" %con)#writing all the required results line by line
    i.write("\n The report for the dawid skene is \n %s"%rep)#writing the report
    i.write("\n The positive recall value using dawid skene is %s "%posrec)#writing the pos recall value
    i.write("\n The negative recall value using dawid skene is  %s"%negvec)#writing the neg recall value
    i.write("\n The F1 using dawid skene is %s"%f1)#writing the f1 score
    i.write("\n The accuracy using dawid skene is  %s"%acc)#writing the accuracy
    i.write("\n The predicted probability for x train is \n %s "%pred)#writing the predicted probability
```

```python
d= StringIO()#intitialising the stringIO
export_graphviz(dtree, out_file=d,filled=True, rounded=True,special_characters=True)
#doing graphical visualisation for obtaining the decision tree
g= pydotplus.graph_from_dot_data(d.getvalue())  
Image(g.create_png())#desired output
```
