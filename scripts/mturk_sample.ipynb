```python
import pandas as pd#loading the required packages
import numpy as np
```

```python
gold_sample=pd.read_csv('../data/gold_sample.csv',index_col=0)#loading the gold sample dataset
gold_sample.head()#showing the output
```

```python
mturk=pd.read_csv('../data/mturk.csv',index_col=0)#loading the mturk dataset
```

```python
mturk_sample=pd.merge(gold_sample,mturk,on='id',suffixes=('','_y'))
#merging the gold sample and mturk based on id and renaming the _x duplicates as nothing and keeping _y as it is for now
mturk_sample=mturk_sample.drop([col for col in mturk_sample.columns if '_y' in col],axis=1)
#dropping all the columns which has _y in their column name so that all the duplicated columns are removed to get final sample
```

```python
mturk_sample.to_csv('../data/mturk_sample.csv')#storing the mturk sample that has been generated in the data folder as told
mturk_sample.head()#showing the output of the mturk sample.
```
